```python
import numpy as np
import random as r
import matplotlib.pyplot as plt

def irrfahrt(runs, n, p, MC):
    endzustaende = []
    
    for _ in range(MC):
        zustand = 1
        
        for _ in range(runs):
            z = r.random()
            
            if z <= p:
                zustand += 1
            else:
                zustand -= 1
            
            if zustand == 0 or zustand == n:
                break
                
        endzustaende.append(zustand)
    
    return np.mean(endzustaende)

wsk = np.arange(0, 1, 0.01)
mittelwerte = []

for p in wsk:
    mittelwerte.append(irrfahrt(100000, 1000, p, 10000))

plt.plot(wsk, mittelwerte)
plt.xlabel("p")
plt.ylabel("mittlerer Endzustand")
plt.show()
```

```python
n_arange = np.arange(0,1000,1)
def symmetrische_irrfahrt (n,start):
    steps=[]
    step=start
    for i in range (n):
        u=r.random()
        if 0<=u<=0.5:
            step -= 1
        elif 0.5<u<=1:
            step +=1
        steps.append(step)
    return steps
steps_drin= (symmetrische_irrfahrt(1000,10))
plt.plot(n_arange,steps_drin)
plt.show()
```
